## Carga y visualización de datos

```python
ind_3 = pd.read_csv('Datos Reloj 3 - Hoja 1.csv')
```

```python
ind_3.head()
```

```python
ind_3 = ind_3.drop(columns='Unnamed: 10')
```

```python
ind_3.describe()
```

```python
def convertir_a_segundos(tiempo):
    horas = tiempo // 100
    minutos = tiempo % 100
    segundos = horas * 3600 + minutos * 60
    return segundos
```

```python
for i in ['Tiempo en la cama', 'Despierto', 'REM', 'Ligero', 'Profundo', 'Durmiendo', 'Conciliacion']:
    ind_3[i] = convertir_a_segundos(ind_3[i])
```

```python
ind_3.head()
```

```python
ind_3.rename(columns={'Tiempo en la cama': 'Tiempo en la cama (s)'}, inplace=True)
ind_3.rename(columns={'Despierto': 'Despierto (s)'}, inplace=True)
ind_3.rename(columns={'REM': 'REM (s)'}, inplace=True)
ind_3.rename(columns={'Ligero': 'Ligero (s)'}, inplace=True)
ind_3.rename(columns={'Profundo': 'Profundo (s)'}, inplace=True)
ind_3.rename(columns={'Durmiendo': 'Durmiendo (s)'}, inplace=True)
ind_3.rename(columns={'Conciliacion': 'Conciliación (s)'}, inplace=True)
```

```python
# Guardamos los datos
ind_3.to_csv('df_sueño_pasos_3.csv')
```
